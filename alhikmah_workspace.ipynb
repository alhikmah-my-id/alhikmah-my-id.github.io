### 2. Kloning & Pembaruan Repositori Git
Gunakan modul `git` untuk mengunduh kode sumber situs dari GitHub secara lokal.

```python
import os
from git import Repo

REPO_URL = "https://github.com/alhikmah-my-id/alhikmah-my-id.github.io.git"
LOCAL_DIR = "./alhikmah_repo"

if not os.path.exists(LOCAL_DIR):
    print(f"Mengunduh repositori ke {LOCAL_DIR}...")
    Repo.clone_from(REPO_URL, LOCAL_DIR)
    print("Kloning Sukses!")
else:
    print("Repositori sudah ada. Melakukan git pull...")
    repo = Repo(LOCAL_DIR)
    repo.remotes.origin.pull()
    print("Repositori berhasil diperbarui!")
```

### 3. Membaca Konfigurasi TOML
Skrip ini membaca parameter internal dari file `config.toml` yang sudah kita buat sebelumnya.

```python
import toml

config_path = "config.toml"

if os.path.exists(config_path):
    config_data = toml.load(config_path)
    print("--- Profil Konfigurasi Situs ---")
    print("Nama Situs:", config_data["site"]["title"])
    print("URL Target:", config_data["site"]["baseURL"])
    print("Tim Pengembang:", config_data["author"]["name"])
else:
    print(f"File {config_path} tidak ditemukan di direktori saat ini.")
```

### 4. Ekstraksi Konten Live (Web Scraping)
Mengambil judul halaman utama dan mengumpulkan daftar link aktif langsung dari domain produksi.

```python
import requests
from bs4 import BeautifulSoup

target_url = "https://alhikmah.my.id/"
headers = {'User-Agent': 'Mozilla/5.0 (Jupyter Notebook Workspace)'}

response = requests.get(target_url, headers=headers)
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    print(f"[Koneksi Berhasil] Judul Situs: {soup.title.text}\n")
    
    print("Mengekstrak 10 Link Pertama:")
    links = soup.find_all('a')
    for i, link in enumerate(links[:10], 1):
        href = link.get('href')
        text = link.text.strip() or "[Tanpa Teks]"
        if href:
            print(f"{i}. {text} -> {href}")
else:
    print(f"Gagal memuat situs. Kode Status: {response.status_code}")
```
